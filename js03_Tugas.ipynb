```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/dataset/spam.csv', encoding='latin-1') # spesifiksi encoding diperlukan karena data tidak menggunakan UTF-8

df.head()
```

```python
# Drop 3 kolom terakhir dengan fungsi iloc
df = df.drop(df.iloc[:,2:], axis=1)

# Cek data
df.head()
```

```python
# Data untuk rename kolom v1 dan v2
new_cols = {
    'v1': 'Labels',
    'v2': 'SMS'
}

# Rename nama kolom v1 dan v2
df = df.rename(columns=new_cols)

# cek data
df.head()
```

```python
# Cek Jumlah Data Per Kelas
print(df['Labels'].value_counts())
print('\n')

# Cek Kelengkapan Data
print(df.info())
print('\n')

# Cek Statistik Deskriptif
print(df.describe())
```

```python
# Data untuk label
new_labels = {
    'spam': 1,
    'ham': 0
}

# Encode label
df['Labels'] = df['Labels'].map(new_labels)

# Cek data
df.head()
```

```python
X = df['SMS'].values
y = df['Labels'].values
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

# Split data training dan testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

# Inisiasi CountVectorizer
bow = CountVectorizer()

# Fitting dan transform X_train dengan CountVectorizer
X_train = bow.fit_transform(X_train)

# Transform X_test
# Mengapa hanya transform? Alasan yang sama dengan kasus pada percobaan ke-3
# Kita tidak menginginkan model mengetahui paramter yang digunakan oleh CountVectorizer untuk fitting data X_train
# Sehingga, data testing dapat tetap menjadi data yang asing bagi model nantinya
X_test = bow.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Inisiasi MultinomialNB
mnb = MultinomialNB()

# Fit model
mnb.fit(X_train, y_train)

# Prediksi dengan data training
y_pred_train = mnb.predict(X_train)

# Evaluasi akurasi data training
acc_train = accuracy_score(y_train, y_pred_train)

# Prediksi dengan data training
y_pred_test = mnb.predict(X_test)

# Evaluasi akurasi data training
acc_test = accuracy_score(y_test, y_pred_test)

# Print hasil evaluasi
print(f'Hasil akurasi data train: {acc_train}')
print(f'Hasil akurasi data test: {acc_test}')
```

Menggunakan stop word

```python
X_train_stop, X_test_stop, y_train_stop, y_test_stop = train_test_split(X, y, test_size=0.2, random_state=50)
bow_stop = CountVectorizer(stop_words=["i","am","you","are","they","we","remove","using","with","is","and"])
X_train_stop = bow_stop.fit_transform(X_train_stop)
X_test_stop = bow_stop.transform(X_test_stop)
# Fit model stop words
mnb.fit(X_train_stop, y_train_stop)

# Prediksi dengan data training stop words
y_pred_train_stop = mnb.predict(X_train_stop)

# Evaluasi akurasi data training stop words
acc_train_stop = accuracy_score(y_train_stop, y_pred_train_stop)

# Prediksi dengan data training stop words
y_pred_test_stop = mnb.predict(X_test_stop)

# Evaluasi akurasi data training stop words
acc_test_stop = accuracy_score(y_test, y_pred_test_stop)

# Print hasil evaluasi stop words
print()
print(f'Hasil akurasi data train dengan stop word: {acc_train_stop}')
print(f'Hasil akurasi data test dengan stop word: {acc_test_stop}')
```

Perbedaan menggunakan stop word
- Hasil akurasi data train: 0.9946152120260264
- Hasil akurasi data train dengan stop word: 0.994166479694862
- Kesimpulan: hasil akurasi data train menggunakan stop word lebih turun daripada normal


- Hasil akurasi data test: 0.9775784753363229
- Hasil akurasi data test dengan stop word: 0.97847533632287
- Kesimpulan: hasil akurasi data test menggunakan stop word berbanding terbalik dengan hasil akurasi data train menggunakan stop word. Dimana menunjukkan nilai yang lebih besar daripada normal
