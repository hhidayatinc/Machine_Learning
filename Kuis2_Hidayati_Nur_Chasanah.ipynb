# Visualisasi Data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(data['tahun'])
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='tahun',hue='skema', data=data)
plt.title('Jumlah skema setiap tahunnya', fontsize=14)
plt.ylabel('Jumlah', fontsize=14)
plt.xlabel('Tahun', fontsize=14)
plt.xticks(rotation='vertical');
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='tahun',hue='luaran_hki', data=data)
plt.title('Ketersediaan luaran HKI tiap tahunnya', fontsize=14)
plt.ylabel('Jumlah', fontsize=14)
plt.xlabel('Tahun', fontsize=14)
plt.xticks(rotation='vertical');
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='tahun',hue='jenis_luaran_lain', data=data)
plt.title('Distribusi tahun ', fontsize=14)
plt.ylabel('Jumlah', fontsize=14)
plt.xlabel('Tahun', fontsize=14)
plt.xticks(rotation='vertical');
```

```python
sns.FacetGrid(data, hue="jenis_luaran_lain", size=6).map(sns.kdeplot, "tahun").add_legend()
```

```python
sns.FacetGrid(data, hue="skema", size=6).map(sns.kdeplot, "tahun").add_legend()
```

```python
sns.heatmap(data.corr(), cmap="YlGnBu", annot = True)
plt.show()
```
